```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Load dataset
dataset = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
# Independent variables
X = dataset[['Age', 'EstimatedSalary']]

# Dependent variable
y = dataset['Purchased']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)

# Create Decision Tree model
classifier = DecisionTreeClassifier(
    criterion='entropy',
    random_state=0
)

# Train model
classifier.fit(X_train, y_train)

# Predict test set results
y_pred = classifier.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)
```

```python
import pandas as pd

# Read CSV file
df = pd.read_csv("employees.csv")

print("Employee Late Arrival Report")
print("-" * 40)

for index, row in df.iterrows():

    late_minutes = row['MinutesLate']

    # Determine reason and penalty
    if late_minutes <= 10:
        reason = "Traffic"
        penalty = 0

    elif late_minutes <= 30:
        reason = "Overslept"
        penalty = 100

    else:
        reason = "Personal Delay"
        penalty = 300

    print(f"\nEmployee ID: {row['EmployeeID']}")
    print(f"Name: {row['EmployeeName']}")
    print(f"Late Minutes: {late_minutes}")
    print(f"Possible Reason: {reason}")
    print(f"Penalty: ₹{penalty}")
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Load datasets
train_data = pd.read_csv("train_data.csv")
test_data = pd.read_csv("test_data.csv")

# Features (X) and Target (y)
X_train = train_data.drop(columns=["Survived", "Unnamed: 0", "PassengerId"])
y_train = train_data["Survived"]

X_test = test_data.drop(columns=["Survived", "Unnamed: 0", "PassengerId"])
y_test = test_data["Survived"]

# Create Decision Tree
classifier = DecisionTreeClassifier(
    criterion="entropy",
    random_state=0
)

# Train model
classifier.fit(X_train, y_train)

# Predict
y_pred = classifier.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

# Load datasets
train_data = pd.read_csv("train_data.csv")
test_data = pd.read_csv("test_data.csv")

# Features and target
X_train = train_data.drop(columns=["Survived", "Unnamed: 0", "PassengerId"])
y_train = train_data["Survived"]

X_test = test_data.drop(columns=["Survived", "Unnamed: 0", "PassengerId"])
y_test = test_data["Survived"]

# Create Random Forest model
classifier = RandomForestClassifier(
    n_estimators=100,
    criterion="entropy",
    random_state=0
)

# Train model
classifier.fit(X_train, y_train)

# Predict
y_pred = classifier.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

cm_display = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, y_pred))
cm_display.plot(cmap='Blues')
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Load datasets
train_data = pd.read_csv("train_data.csv")
test_data = pd.read_csv("test_data.csv")

# Features and target
X_train = train_data.drop(columns=["Survived", "Unnamed: 0", "PassengerId"])
y_train = train_data["Survived"]

X_test = test_data.drop(columns=["Survived", "Unnamed: 0", "PassengerId"])
y_test = test_data["Survived"]

# Feature Scaling
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Logistic Regression Model
classifier = LogisticRegression(random_state=0)

# Train model
classifier.fit(X_train, y_train)

# Predict
y_pred = classifier.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load data
train_data = pd.read_csv("train_data.csv")
test_data = pd.read_csv("test_data.csv")

# Convert Sex
train_data["Sex"] = train_data["Sex"].map({"male": 0, "female": 1})
test_data["Sex"] = test_data["Sex"].map({"male": 0, "female": 1})

# Impute missing 'Age' values with the mean
train_data["Age"].fillna(train_data["Age"].mean(), inplace=True)
test_data["Age"].fillna(test_data["Age"].mean(), inplace=True)

# Impute missing 'Sex' values with the mode
train_data["Sex"].fillna(train_data["Sex"].mode()[0], inplace=True)
test_data["Sex"].fillna(test_data["Sex"].mode()[0], inplace=True)

# Features
X_train = train_data[["PassengerId", "Sex", "Age", "Fare"]]
y_train = train_data["Survived"]

X_test = test_data[["PassengerId", "Sex", "Age", "Fare"]]
y_test = test_data["Survived"]

# Feature Scaling
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Logistic Regression
classifier = LogisticRegression(random_state=0)

classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```
