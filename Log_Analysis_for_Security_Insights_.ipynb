# Define a Dictionary to display the Protocol Name

```python
# Dictionary mapping protocol numbers to names
protocol_names = {
    1: "ICMP",
    6: "TCP",
    17: "UDP"
    # Add more protocols as needed
}
```

# Determine the type of file based on its extension

```python
def parse_file(file_path):
    parsed_data = []
    if file_path.endswith('.pcap'):
        parsed_data = parse_pcap(file_path)
        file_type = "PCAP"
    else:
        print("Unsupported file type.")
        return None, None

    return parsed_data, file_type
```

# Extract Information

```python
def parse_pcap(pcap_file):
    parsed_packets = []
    packets = rdpcap(pcap_file)
    for pkt in packets:
        try:
            # Extracting relevant information from each packet
            ip_src = pkt[IP].src
            ip_dst = pkt[IP].dst
            protocol_num = pkt[IP].proto
            protocol_name = protocol_names.get(protocol_num, "Unknown")
            length = len(pkt)

            # Constructing a dictionary for each packet
            packet_info = {
                "source_ip": ip_src,
                "destination_ip": ip_dst,
                "protocol": protocol_name,
                "length": length
            }
            parsed_packets.append(packet_info)
        except IndexError:
            # Handling any incomplete or invalid packets
            continue
    return parsed_packets
```

# Function to Search for a given query string

```python
def search(parsed_data, search_query):
    results = []
    for item in parsed_data:
        if isinstance(item, dict):
            # Searching packet data
            if any(search_query in str(value) for value in item.values()):
                results.append(item)
        elif isinstance(item, str):
            # Searching log entries
            if search_query in item:
                results.append(item)
    return results
```

# Feed the Log file as Input

```python
pcap_file_path = 'sample.pcap'
parsed_packets = parse_pcap(pcap_file_path)
for pkt in parsed_packets:
    print(pkt)
```

# Search Query

```python
# Example usage:
file_path = 'sample.pcap'
parsed_data, file_type = parse_file(file_path)
if parsed_data and file_type:
    # Perform searching
    search_query = "192.168.1.1"  # Example search query
    search_results = search(parsed_data, search_query)

    # Display results
    print(f"File Type: {file_type}")
    print(f"Search Results for '{search_query}':")
    for result in search_results:
        print(result)
```
